```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# %matplotlib notebook
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r') #red

plt.subplot(1,2,2)
plt.plot(y,x,'b') #blue
```

```python
# Object Oriented
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
#for current_ax in axes:
#    current_ax.plot(x,y)
axes[0].plot(x,y,color='purple',linewidth=2,alpha=0.5,linestyle='--',marker='o')
axes[0].set_title('First Plot')
axes[1].plot(y,x,linestyle='-.')
#axes[1].set_xlim([0,1])
#axes[1].set_ylim([0,2])
axes[1].set_title('Second Plot')
plt.tight_layout()
```

### Figure size and DPI

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
# fig.savefig('my_pic.png',dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')
ax.legend()
```

### Visualizing Time Series

```python
mcdon = pd.read_csv('mcdonalds.csv',index_col='Date',parse_dates=True)
mcdon.head()
```

```python
mcdon.plot()
```

```python
mcdon['Adj. Close'].plot()
```

```python
mcdon['Adj. Volume'].plot(figsize=(12,4))
```

```python
mcdon['Adj. Close'].plot(xlim=['2007-01-01','2009-01-01'],ylim=(20,50),ls='--',c='red')
```

```python
import matplotlib.dates as dates
```

```python
idx = mcdon.loc['2007-01-01':'2007-05-01'].index
idx
```

```python
stock = mcdon.loc['2007-01-01':'2007-05-01']['Adj. Close']
stock
```

```python
fig,ax = plt.subplots()
ax.plot_date(idx, stock,'-')
ax.xaxis.grid(True)
ax.yaxis.grid(True)
# LOCATING
ax.xaxis.set_major_locator(dates.MonthLocator())
# FORMATING
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n%b-%Y'))
ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))
fig.autofmt_xdate()
plt.tight_layout()
```
