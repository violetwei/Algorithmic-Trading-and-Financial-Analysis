### Hedging

Recall that Beta represents an asset's exposure to the market (typically represented as the S&P500)

Trading strategies with a low Beta are attractive, as they should operate independently from the market

In the following, we will calculate an asset's Beta and Alpha and then show a simple method of hedging againset the Beta, in an attempt to cancel out any risk exposure to the market

```python
import numpy as np
from statsmodels import regression
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
start = '2019-07-01'
end = '2020-07-30'
```

```python
asset = get_pricing('AAPL', fields='price',start_date=start,end_date = end)
```

```python
benchmark = get_pricing('SPY', fields='price',start_date=start,end_date=end)
```

```python
asset_ret = asset.pct_change(1)[1:]
bench_ret = benchmark.pct_change(1)[1:]
```

```python
asset_ret.plot()
bench_ret.plot()
plt.legend()
```

```python
plt.scatter(bench_ret,asset_ret,alpha=0.6,s=50)
plt.xlabel('SPY Return')
plt.ylabel('AAPL Return')
```

```python
AAPL = asset_ret.values
spy = bench_ret.values
```

```python
spy_constant = sm.add_constant(spy)
```

```python
model = regression.linear_model.OLS(AAPL,spy_constant).fit()
```

```python
model.params
```

```python
alpha, beta = model.params
```

```python
min_spy = bench_ret.values.min()
max_spy = bench_ret.values.max()

spy_line = np.linspace(min_spy,max_spy,100)
y = spy_line*beta + alpha

plt.plot(spy_line,y,'r')
plt.scatter(bench_ret,asset_ret,alpha=0.6,s=50)
plt.xlabel('SPY Return')
plt.ylabel('AAPL Return')
```

```python
hedged = -1*(beta*bench_ret) + asset_ret
```

```python
hedged.plot(label='AAPL with Hedge')
asset_ret.plot(alpha=0.5)
bench_ret.plot(alpha=0.5)
plt.legend()
```

```python
hedged.plot(label='AAPL with Hedge')
asset_ret.plot(alpha=0.5)
bench_ret.plot(alpha=0.5)
plt.xlim(['2020-2-15','2020-05-15'])
plt.legend()
```

```python
def alpha_beta(benchmark_ret,stock):
    
    benchmark = sm.add_constant(benchmark_ret)
    
    model = regression.linear_model.OLS(stock,benchmark).fit()
    
    return model.params[0],model.params[1]
```

```python
start = '2019-07-01'
end = '2020-07-30'
asset2019 = get_pricing('AAPL', fields='price',start_date=start, end_date=end)
benchmark2019 = get_pricing('SPY', fields='price',start_date=start,end_date=end)

asset_ret2019=asset2019.pct_change(1)[1:]
benchmark_ret2019=benchmark2019.pct_change(1)[1:]

aret_values = asset_ret2019.values
bret_values = benchmark_ret2019.values

alpha2019,beta2019 = alpha_beta(bret_values,aret_values)

print('2019 VALUES')
print('AAPL alpha '+str(alpha2019))
print('AAPL beta '+str(beta2019))
```

```python
portfolio = -1*beta2019*benchmark_ret2019 + asset_ret2019
```

```python
alpha,beta = alpha_beta(benchmark_ret2019,portfolio)
```

```python
print('PORTFOLIO ALPHA AND BETA')
print('alpha '+str(alpha))
print('beta '+str(beta))
```

```python
portfolio.plot(alpha=0.9,label='AAPL with HEDGE')
asset_ret2019.plot(alpha=0.5)
benchmark_ret2019.plot(alpha=0.5)
plt.ylabel('DAILY RETURN')
plt.legend()
```
